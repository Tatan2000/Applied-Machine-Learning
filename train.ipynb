```python
#Importing necessary packages
import numpy as np
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import seaborn as sns
```

```python
#importing data
train = pd.read_csv(r'train.csv')
test = pd.read_csv(r'test.csv')
validation = pd.read_csv(r'validation.csv')
```

```python
X = train['message']
y = train['label']
X_validation = validation['message']
y_validation = validation['label']
X_test = test['message']
y_test = test['label']
```

## Multinomial Naive-Bayes

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X,y)
predicted_categories_validation = model.predict(X_validation)
predicted_categories_train = model.predict(X)
```

### Training data


```python
cm_nb_tr = confusion_matrix(y, predicted_categories_train)
ax = plt.subplot()
sns.heatmap(cm_nb_tr, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Naive-bayes: Confusion Matrix for training data'); 
ax.xaxis.set_ticklabels(['ham','spam']); ax.yaxis.set_ticklabels(['ham', 'spam']);
print("The training accuracy is {}".format(accuracy_score(y, predicted_categories_train)))

print("Precision score for training data:", precision_score(y, predicted_categories_train))
print("Recall Score for training data:", recall_score(y, predicted_categories_train))
```

### Validation data

```python
cm_nb_va = confusion_matrix(y_validation, predicted_categories_validation)
ax = plt.subplot()
sns.heatmap(cm_nb_va, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Naive-bayes: Confusion Matrix for validation data'); 
ax.xaxis.set_ticklabels(['ham','spam']); ax.yaxis.set_ticklabels(['ham', 'spam']);
print("The validation accuracy is {}".format(accuracy_score(y_validation, predicted_categories_validation)))

print("Precision score for validation data:", precision_score(y_validation, predicted_categories_validation))
print("Recall Score for validation data:", recall_score(y_validation, predicted_categories_validation))
```

## Support Vector Classifier(SVC)

```python
model_2 = make_pipeline(TfidfVectorizer(),SVC(kernel='linear', random_state=0))
model_2.fit(X,y) 
predicted_categories_validation = model_2.predict(X_validation)
predicted_categories_train = model_2.predict(X)
```

### Training data

```python
cm_svc_tr = confusion_matrix(y, predicted_categories_train)
ax = plt.subplot()
sns.heatmap(cm_svc_tr, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Support Vector Classifier: Confusion Matrix for training data'); 
ax.xaxis.set_ticklabels(['ham','spam']); ax.yaxis.set_ticklabels(['ham', 'spam']);
print("The training accuracy is {}".format(accuracy_score(y, predicted_categories_train)))

print("Precision score for training data:", precision_score(y, predicted_categories_train))
print("Recall Score for training data:", recall_score(y, predicted_categories_train))
```

### Validation data

```python
cm_svc_va = confusion_matrix(y_validation, predicted_categories_validation)
ax = plt.subplot()
sns.heatmap(cm_svc_va, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Singular Value Classifier: Confusion Matrix for validation data'); 
ax.xaxis.set_ticklabels(['ham','spam']); ax.yaxis.set_ticklabels(['ham', 'spam']);
print("The validation accuracy is {}".format(accuracy_score(y_validation, predicted_categories_validation)))

print("Precision score for validation data:", precision_score(y_validation, predicted_categories_validation))
print("Recall Score for validation data:", recall_score(y_validation, predicted_categories_validation))
```

## Logistic Regression

### Validation data

```python
cm_logr_va = confusion_matrix(y_validation, predicted_categories_validation)
ax = plt.subplot()
sns.heatmap(cm_logr_va, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Logistic Regression: Confusion Matrix for validation data'); 
ax.xaxis.set_ticklabels(['ham','spam']); ax.yaxis.set_ticklabels(['ham', 'spam']);
print("The validation accuracy is {}".format(accuracy_score(y_validation, predicted_categories_validation)))

print("Precision score for validation data:", precision_score(y_validation, predicted_categories_validation))
print("Recall Score for validation data:", recall_score(y_validation, predicted_categories_validation))
```
