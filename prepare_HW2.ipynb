```python
import pandas as pd
file_path = 'E:\Assigments\MSc Semester 2\AML - Raghav\Programming assignmnet 2\SMSSpamCollection.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

def split(df, train_size = 0.7, validation_size = 0.15, test_size = 0.15, random_state=42):
    df_train, df_ = train_test_split(df, train_size = train_size, test_size = (validation_size + test_size), random_state = random_state)
    df_validation, df_test = train_test_split(df_, train_size = 0.5, test_size = 0.5)
    
    return df_train, df_validation, df_test

df_train, df_validation, df_test = split(df)
```

```python
train_file_path = 'train.csv'
validation_file_path = 'validation.csv'
test_file_path = 'test.csv'

df_train.to_csv(train_file_path, index=False)
df_validation.to_csv(validation_file_path, index=False)
df_test.to_csv(test_file_path, index=False)
```
